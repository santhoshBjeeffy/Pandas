```
csv_url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/Students_Alcohol_Consumption/student-mat.csv'
df = pd.read_csv(csv_url)
df.head()
```

# For the purpose of this exercise slice the dataframe from 'school' until the 'guardian' column

```
stud_alcohol=df.loc[:,"school":"guardian"]
stud_alcohol
```

#create a lambda function that capitalize the strings

```
capitalize=lambda X: X.capitalize()
```

```
#capitalize both the Mjob and Fjob
stud_alcohol['Mjob'].apply(capitalize)
stud_alcohol['Fjob'].apply(capitalize)
```

```
#print the last elements in the dataset
stud_alcohol.tail()
```

```
#we observe that still the original dataframe is not capitalized
stud_alcohol['Mjob'] = stud_alcohol['Mjob'].apply(capitalize)
stud_alcohol['Fjob'] = stud_alcohol['Fjob'].apply(capitalize)
```

#Create a function called majority that return a boolean value to a new column called legal_drinker (Consider majority as older than 17 years old

```
def majority(X):
  if X > 20:
    return True
  else:
    return False
```

```
stud_alcohol['legal_drinker'] = stud_alcohol['age'].apply(majority)
```
